```python
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
type(iris)
```

```python
import pandas as pd
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
print(iris.DESCR)
```

```python
x=iris.data
```

```python
x
```

```python
y = iris.target
```

```python
y
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y = to_categorical(y)
y.shape
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
x_train
```

```python
x_test
y_train
y_test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar_object = MinMaxScaler()
scalar_object.fit(x_train)
```

```python
scaled_x_train = scalar_object.transform(x_train)
scalar_x_test = scalar_object.transform(x_test)
```

```python
x_train.max()
```

```python
scaled_x_train.max()
```

```python
x_train
```

```python
scaled_x_train
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
import tensorflow as tf
```

```python
model =Sequential()
model.add(Dense(0, input_dim=4, activation='relu'))
model.add(Dense(6, input_dim=4, activation='relu'))
model.add(Dense(3,activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(scaled_x_train, y_train, epochs=150, verbose=2)
```

```python
scalar_x_test
```

```python
model.predict(scalar_x_test)
```

```python
model.metrics_names
```

```python
model.evaluate(x=scalar_x_test,y= y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
predictions =(model.predict(x_test) > 0.5).astype("int32")
```

```python
predictions
```

```python
y_test.argmax(axis=1)
```

```python
confusion_matrix(y_test.argmax(axis=1), predictions.argmax(axis=1))
```

```python
print(classification_report(y_test.argmax(axis=1), predictions.argmax(axis=1)))
```
