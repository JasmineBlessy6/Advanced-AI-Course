```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
doc1 = nlp('I am runner running in a race because I love to run since I ran today')
for token in doc1:
  print(token.text,'\t', token.pos_ ,'\t', token.lemma ,'\t',token.lemma_)
```

```python
def show_lemmas(text):
  for token in text:
    print(f'{token.text:{12}} {token.pos_:{6}} { token.lemma:{22}} {token.lemma_}')
```

```python
doc2 = nlp(u"I saw a eighteen mice today!")
show_lemmas(doc2)
```

```python
doc3 = nlp(u'I am meeting him tomorrow at the meeting')
show_lemmas(doc3)
```
