```python
def show_ents(doc):
  if doc.ents:
    for ent in doc.ents:
      print(ent.text+'-'+ent.label_+'-'+ str(spacy.explain(ent.label_)))
  else:
    print('No named entities found')
```

```python
doc = nlp(u'May I go to Washington, DC next May to see the Washington Monuments?')
show_ents(doc)
```

```python
doc = nlp(u'Can I please borrow 500 dollars from you to buy some microsoft stack?')
for ent in doc.ents:
  print(ent.text,ent.start,ent.end)
```

```python
doc = nlp(u'Tesla to build a U.K factory for $6 million')
show_ents(doc)
```

```python
from spacy.matcher import PhraseMatcher
matcher = PhraseMatcher(nlp.vocab)
```

```python
doc = nlp(U"Our company plans to introduce a new vaccum cleaner"
          U" If Successful the vaccum cleaner will be our first product")
Phrase_list = {'vaccum cleaner','vaccum_cleaner'}
phrase_patterns = [nlp(text) for text in Phrase_list]
matcher.add('newProduct', None , *phrase_patterns)
matches = matcher(doc)
matches
```
