```python
# Install required libraries
!pip install nltk PyPDF2

import nltk

from nltk.stem import PorterStemmer, SnowballStemmer, LancasterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import PyPDF2
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
# Download necessary NLTK data
#nltk.download('punkt')
nltk.download('wordnet')
#nltk.download('omw-1.4')
nltk.download('stopwords')
```

```python
text = "He has obstructed the Administration of Justice, by refusing his Assent to Laws for establishing Judiciary powers."
```

```python
doc = nlp(text)
```

```python
text
```

```python
# TOKENIZATION
# 1. Simple word_tokenize
for token in doc:
    print(token.text)

```

```python
print("\nToken Attributes:")
for token in doc:
    print(f"Token: {token.text}, Lemma: {token.lemma_}, POS: {token.pos_}, Is Stop: {token.is_stop}")
```

```python
# STEMMING
porter = PorterStemmer()
tokens1 = [token.text for token in doc]
porter_stems = [porter.stem(word) for word in tokens1]
print("Porter Stemmer:", porter_stems)
```

```python
# 2. Snowball Stemmer
snowball = SnowballStemmer('english')
snowball_stems = [snowball.stem(word) for word in tokens1]
print("Snowball Stemmer:", snowball_stems)
```

```python
#LEMMATISATION
def show_lemmas(text):
  for token in text:
    print(f'{token.text:{12}} {token.pos_:{6}} { token.lemma:{22}} {token.lemma_}')
```

```python
show_lemmas(doc)
```
