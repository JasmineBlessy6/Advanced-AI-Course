```python
import spacy
from spacy import displacy
nlp = spacy.load("en_core_web_sm")
```

```python
sentence = "He has obstructed the Administration of Justice, by refusing his Assent to Laws for establishing Judiciary powers."
```

```python
doc = nlp(sentence)
```

```python
# Entities in sentence
print("Entities in the sentence:")
for ent in doc.ents:
    print(f"{ent.text} ({ent.label_})")
```

```python
# Identify noun chunks
print("\nNoun chunks:")
for chunk in doc.noun_chunks:
    print(f"{chunk.text}")
```

```python
#iii) Draw the displacy diagram
spans = list(doc.sents)
options = {'distance':110,'compact':False,'color':'white','bg':'black','font':'Times'}
displacy.render(doc, style='dep',jupyter=True,options=options)
```

```python
#Identify the pattern
print("\nDependency patterns:\n")
for token in doc:
  print(f"{token.text:{15}}{token.pos_:{15}}{token.tag_:{10}}{spacy.explain(token.tag_)}")
```
