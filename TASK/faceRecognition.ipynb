```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
prasinna= cv2.imread('/content/drive/MyDrive/VAC-PROJECT/prasinna.jpg',0)
blessy = cv2.imread('/content/drive/MyDrive/VAC-PROJECT/blessy.png',0)
panjanadhan= cv2.imread('/content/drive/MyDrive/VAC-PROJECT/panjanadhan.jpg',0)
jency= cv2.imread('/content/drive/MyDrive/VAC-PROJECT/jency.png',0)
jeffrin= cv2.imread('/content/drive/MyDrive/VAC-PROJECT/jeffrin.png',0)
lazarus= cv2.imread('/content/drive/MyDrive/VAC-PROJECT/lazarus.jpg',0)
group = cv2.imread('/content/drive/MyDrive/VAC-PROJECT/group.png',0)
richan = cv2.imread('/content/richan.png',0)
```

```python
face_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/Advanced AI/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
  face_img = img.copy()
  face_rects = face_cascade.detectMultiScale(face_img)
  for(x,y,w,h) in face_rects:
    cv2.rectangle(face_img,(x,y), (x+w,y+h), (255,255,255), 5)
  return face_img
```

```python
result = detect_face(richan)
plt.imshow(result, cmap='gray')
plt.show()
```

```python
def adj_detect_face(img):
  face_img = img.copy()
  face_rects = face_cascade.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=5)
  for(x,y,w,h) in face_rects:
    cv2.rectangle(face_img,(x,y), (x+w,y+h), (255,255,255), 10)
  return face_img
```

```python
result = adj_detect_face(group)
plt.imshow(result, cmap='gray')
plt.show()
```
