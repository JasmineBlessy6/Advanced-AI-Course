```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
blank_img = np.zeros((512,512,3), dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
plt.show
```

```python
cv2.rectangle(blank_img, pt1=(384,0), pt2=(510,234), color = (0,255,0), thickness=5)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.rectangle(blank_img, pt1=(200,200), pt2=(300,300), color = (0,0,255), thickness=5)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.circle(blank_img, center=(250,250), radius=40,color=(255,0,0), thickness=5)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.line(blank_img, pt1=(0,0), pt2=(511,511),color=(102,255,255), thickness=5)
plt.imshow(blank_img)
plt.show()
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text='Hello', org=(10,500), fontFace=font, fontScale=4, color=(255,255,255), thickness=2, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
blank_img = np.zeros((512,512,3), dtype=np.int32)
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],np.int32)
```

```python
pts = vertices.reshape((-1,1,2))
pts
```

```python
cv2.polylines(blank_img, [pts], isClosed=True, color=(255,0,0),thickness=5)
plt.imshow(blank_img)
```
